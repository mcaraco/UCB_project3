```python
QII = pd.read_csv('ISO_Numbeo/QII.csv')
table_QII = QII.drop(columns=['Unnamed: 0'])
table_QII.head()
```

```python
# Union by Year for Plotly Scatter Graph
# Need to merge Continent AND Population Data
```

```python
# Read in Contients CSV
df_cont = pd.read_csv('https://pkgstore.datahub.io/JohnSnowLabs/country-and-continent-codes-list/country-and-continent-codes-list-csv_csv/data/b7876b7f496677669644f3d1069d3121/country-and-continent-codes-list-csv_csv.csv')
df_cont.head()
```

```python
# rename columns
df_cont = df_cont[['Continent_Name','Three_Letter_Country_Code']].rename(columns={
    'Three_Letter_Country_Code':'iso_a3',
    'Continent_Name':'Continent'
})
df_cont.head()
```

```python
# Read in Population CSV
df_pop = pd.read_csv('https://pkgstore.datahub.io/core/population/population_csv/data/ead5be05591360d33ad1a37382f8f8b1/population_csv.csv')
df_pop.head()
```

```python
# rename columns
df_pop1 = df_pop[['Country Code','Year','Value']].rename(columns={
    'Country Code':'iso_a3',
    'Year':'year',
    'Value':'pop'
}).copy()
df_pop1.head()
```

```python
# plotly will only have years 2012-2016 since population data only has upto 2016
df_pop1['year'].max()
```

```python
# Merge continents using iso_a3
df_t = Df1.merge(df_cont,how='left',on='iso_a3')
df_t.head()
```

```python
# For 2012
df_2012 = df_t[['Country','QII_2012','GDP_2012','iso_a3','Continent']].copy().rename(columns={
    'QII_2012':'QII',
    'GDP_2012':'GDP'
})
df_2012['year'] = 2012
df_2012.head()
```

```python
# For 2013
df_2013 = df_t[['Country','QII_2013','GDP_2013','iso_a3','Continent']].copy().rename(columns={
    'QII_2013':'QII',
    'GDP_2013':'GDP'
})
df_2013['year'] = 2013
df_2013.head()
```

```python
# For 2014
df_2014 = df_t[['Country','QII_2014','GDP_2014','iso_a3','Continent']].copy().rename(columns={
    'QII_2014':'QII',
    'GDP_2014':'GDP'
})
df_2014['year'] = 2014
df_2014.head()
```

```python
# For 2015
df_2015 = df_t[['Country','QII_2015','GDP_2015','iso_a3','Continent']].copy().rename(columns={
    'QII_2015':'QII',
    'GDP_2015':'GDP'
})
df_2015['year'] = 2015
df_2015.head()
```

```python
# For 2016
df_2016 = df_t[['Country','QII_2016','GDP_2016','iso_a3','Continent']].copy().rename(columns={
    'QII_2016':'QII',
    'GDP_2016':'GDP'
})
df_2016['year'] = 2016
df_2016.head()
```

```python
# For 2017
df_2017 = df_t[['Country','QII_2017','GDP_2017','iso_a3','Continent']].copy().rename(columns={
    'QII_2017':'QII',
    'GDP_2017':'GDP'
})
df_2017['year'] = 2017
df_2017.head()
```

```python
# Concatenate 2012-2016
df_plotly = pd.concat([df_2012,df_2013,df_2014,df_2015,df_2016])
```

```python
# check
df_plotly.sample(n=10)
```

```python
# Merge population data by iso_a3 and year
df_plotly = df_plotly.merge(df_pop1,how='left',on=['iso_a3','year'])
```

```python
# check
df_plotly[df_plotly['Country']=='United States']
```

```python
# rename to all lower case var names
df_plotly = df_plotly.rename(columns={
    'Country':'country',
    'QII':'qii',
    'GDP':'gdp',
    'Continent':'continent'
}).copy()
df_plotly.head()
```

```python
# export new continent data as CSV
df_plotly.to_csv('plotlyJS/QII_GDP_Year.csv')
```
